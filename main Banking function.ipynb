```python
# Banking system Assignmet

#program 


import random
import string
import os
import json

def generate_account_number():
    """Generate a random 10-digit Account number"""
    return ''.join(random.choices(string.digits, k=10))

def generate_password():
    """Generate a random 8-character Password"""
    return ''.join(random.choices(string.ascii_letters + string.digits, k=8))

def save_account(account):
    """Save account details to a file."""
    with open("accounts.json", "a") as file:
        json.dump(account, file)
        file.write("\n")

def load_accounts():
    """Load all accounts from the file."""
    accounts = []
    if os.path.exists("accounts.json"):
        with open("accounts.json", "r") as file:
            for line in file:
                accounts.append(json.loads(line.strip()))
    return accounts

def create_account(name, initial_balance=0):
    """Create a new account with a random account number and password."""
    account_number = generate_account_number()
    password = generate_password()
    account = {
        "name": name,
        "account_number": account_number,
        "password": password,
        "balance": initial_balance,
        "transactions": []
    }
    save_account(account)
    with open(f"{account_number}_transactions.txt", "w") as file:
        file.write(f"Transaction history for account {account_number}\n")
    print(f"Account created!\nAccount Number: {account_number}\nPassword: {password}")
    return account

def deposit(account, amount):
    """Add money to the Account."""
    if amount > 0:
        account["balance"] += amount
        transaction = f"Deposit: ${amount:.2f}. New Balance: ${account['balance']:.2f}"
        account["transactions"].append(transaction)
        _record_transaction(account["account_number"], transaction)
        print(transaction)
    else:
        print("Deposit amount must be positive.")

def withdraw(account, amount):
    """Withdraw money from the Account."""
    if amount > 0 and amount <= account["balance"]:
        account["balance"] -= amount
        transaction = f"Withdrawal: ${amount:.2f}. New Balance: ${account['balance']:.2f}"
        account["transactions"].append(transaction)
        _record_transaction(account["account_number"], transaction)
        print(transaction)
    elif amount > account["balance"]:
        print("Insufficient balance.")
    else:
        print("Withdrawal amount must be positive.")

def check_balance(account):
    """Display the current Balance."""
    print(f"Current Balance: ${account['balance']:.2f}")

def print_statement(account):
    """Print all transactions from Memory and File."""
    print(f"\nStatement for {account['name']} ({account['account_number']}):")
    for transaction in account["transactions"]:
        print(transaction)
    print("\nTransactions from file:")
    file_name = f"{account['account_number']}_transactions.txt"
    if os.path.exists(file_name):
        with open(file_name, "r") as file:
            for line in file:
                print(line.strip())
    else:
        print("No transactions found.")

def _record_transaction(account_number, transaction):
    """Save a transaction to the file."""
    with open(f"{account_number}_transactions.txt", "a") as file:
        file.write(transaction + "\n")

def main():
    print("""
            ------------------------------
            Welcome to the Banking System!
            ------------------------------""")
    accounts = load_accounts()

    while True:
        print("\nMenu:")
       
        print("1. Create Account")
        
        print("2. Deposit")
        
        print("3. Withdraw")
        
        print("4. Check Balance")
        
        print("5. Print Statement")
        
        print("6. Exit")
        
        choice = input("Choose an option: ")

        if choice == "1":
            name = input("Enter your name: ")
            
            initial_balance = float(input("Enter initial balance: "))
            
            account = create_account(name, initial_balance)
            
            accounts.append(account)
        elif choice in ["2", "3", "4", "5"]:
            account_number = input("Enter account number: ")
            
            password = input("Enter password: ")
            
            account = next((acc for acc in accounts if acc["account_number"] == account_number and acc["password"] == password), None)
            
            if not account:
                print("Account not found or incorrect password.")
                continue

            if choice == "2":
                amount = float(input("Enter deposit amount: "))
                deposit(account, amount)
            elif choice == "3":
                amount = float(input("Enter withdrawal amount: "))
                withdraw(account, amount)
            elif choice == "4":
                check_balance(account)
            elif choice == "5":
                print_statement(account)
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")


```

```python
main()
```
